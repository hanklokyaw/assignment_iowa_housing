```python
import pandas as pd


file = './__datasets/Ames Housing Dataset.xlsx'

housing = pd.read_excel(io=file)

# names = price_per_neighbors["Neighborhood"] == 'NAmes'
# gilbert = price_per_neighbors["Neighborhood"] == 'Gilbert'

price_per_neighbors = housing[["Neighborhood", "Lot_Area", "Sale_Price"]]
# price_per_neighbors.groupby(by="Neighborhood").describe().round(2)
price_per_neighbors["Price_Per_SF"] = round(price_per_neighbors.Sale_Price/price_per_neighbors.Lot_Area,2)


price_per_sf_per_neighbors = price_per_neighbors[["Neighborhood", "Price_Per_SF"]]
price_per_sf_per_neighbors.groupby(by="Neighborhood").describe().round(2)
```

```python
housing_psf = housing.copy()
housing_psf["Price_per_sf"] = round(housing_psf.Sale_Price/housing_psf.Lot_Area,2)
housing_psf
housing_psf[housing_psf["Neighborhood"]=="Blueste"].sort_values("Price_per_sf", ascending=False)
```
